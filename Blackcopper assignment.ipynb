## Web Scraping

```python
# creating a function which can extract the text from the webpage and then retrieve all the paragraphs:

def extract_para(link):

        sourcecode = requests.get(link)

        soup = BeautifulSoup(sourcecode.text,"lxml")

        paras = []

        # Find and saving all paras
        all_paras = soup.find_all('p')

        for para in all_paras:
            para_str = str(para)

            if para_str.startswith('<p>'):
                paras.append(para_str)
                
        return paras
```

```python
df.head()
```

```python
# applying the function on dataframe to add all the paragraph texts to its corresponding URL_ID. 

df['text'] = df['URL'].apply(extract_para)
```

```python
df.head()
```

```python
# creating a function to extract the title of the article from the webpage:

def extract_title(link):

        sourcecode = requests.get(link)

        soup = BeautifulSoup(sourcecode.text,"lxml")

        # Find and save title:
        title_tag = soup.title
        
        empty = []
        
        for i in title_tag.text.split():
            empty.append(i)
                
        return " ".join(empty)
```

```python
# applying the function on the dataframe to save the title of it's corresponding URL_ID:

df['title'] = df['URL'].apply(extract_title)
```

## Data Cleaning

```python
# creating a function which can remove all the speacial characters from the text:

def clean(text):
    # Remove HTML tags and other special characters
    text = str(text)
    cleaned_text = re.sub('<[^<>#]+?>', '', text)
    
    return eval(cleaned_text)
```

```python
df.head()
```

```python
# applying the function on the dataframe:

df['text'] = df['text'].apply(clean)
```

```python
df.head()
```

```python
# creating one more function which can remove all the speacial characters from the text except alphabets and numbers:

def remove_char(list1):
    string = ''
    for i in list1:
#         string += i.replace(',','').replace('#','')
        x = re.sub('[^a-zA-Z0-9.]',' ',i)
        string += x
        
    return string.lower()
```

```python
# applying the function on the dataframe:

df['text'] = df['text'].apply(remove_char)
```

```python
df.head()
```

```python
''' creating a function which can remove the "Blackcoffer Insights" written at the end of each title, and also remove all the 
characters except alphabets and numbers.'''

def edit_title(title):
    title = title.lower()
    
    name = []
    
    for i in title.split():
        if i in ('blackcoffer', 'insights'):
            continue
        else:
            name.append(i)
    
    name = " ".join(name)
    
    x = re.sub('[^a-zA-Z0-9]',' ',name)
    
    return x
```

```python
# applying the function on the dataframe:

df['title'] = df['title'].apply(edit_title)
```

```python
# importing all the text files containing different types of stopwords:

generic = "C:\\Users\\dk957\\Downloads\\drive-download-20231129T114014Z-001\\StopWords_Generic.txt"
auditor = "C:\\Users\\dk957\\Downloads\\drive-download-20231129T114014Z-001\\StopWords_Auditor.txt"
currencies = "C:\\Users\\dk957\\Downloads\\drive-download-20231129T114014Z-001\\StopWords_Currencies.txt"
datesandnumbers = "C:\\Users\\dk957\\Downloads\\drive-download-20231129T114014Z-001\\StopWords_DatesandNumbers.txt"
geographic = "C:\\Users\\dk957\\Downloads\\drive-download-20231129T114014Z-001\\StopWords_Geographic.txt"
names = "C:\\Users\\dk957\\Downloads\\drive-download-20231129T114014Z-001\\StopWords_Names.txt"

# Open the file in read mode
with open(generic, 'r') as file:
    generic_words = file.read().split()
    
with open(auditor, 'r') as file:
    auditor_words = file.read().split()
    
with open(currencies, 'r') as file:
    currencies_words = file.read().split()
    
with open(datesandnumbers, 'r') as file:
    datesandnumbers_words = file.read().split()
    
with open(geographic, 'r') as file:
    geographic_words = file.read().split()
    
with open(names, 'r') as file:
    names_words = file.read().split()
```

```python
# creating a programe which collect all the stopwords from different text files and create a final stopwords list:

stopwords = []

for i in [generic_words,auditor_words,currencies_words,datesandnumbers_words,geographic_words,names_words]:
    for j in i:
        if j=='|':
            continue
        else:
            stopwords.append(j)
            
            
stopwords = [i.lower() for i in stopwords]
stopwords.extend('p, a')
```

```python
# creating a function to remove the stopwords from the given text:

def remove_stopwords(text):

    result = []
    sentences = []

    sentence_list = sent_tokenize(text)

    for sentence in sentence_list:
        words = word_tokenize(sentence)
        filtered_words = [word for word in words if word.lower() not in stopwords]
        result.append(", ".join(filtered_words))
        
    for i in result:
        sentences.append(i.replace(',',' ').replace(".",''))

    return sentences
```

```python
# applying the function on the text feature:

df['text'] = df['text'].apply(remove_stopwords)
```

```python
# creating a function to word tokanize the given list of sentences:

def word_tokanization(list_of_sentences):
    
    words = []
    
    for sentence in list_of_sentences:
        words.append(word_tokenize(sentence))
        
    return words
```

```python
# applying the function on the text feature:
df['text'] = df['text'].apply(word_tokanization)
```

```python
# creating a function to word tokanize and remove stopwords from the title:

def title_token(title):
    
    words =  word_tokenize(title)
    
    words = [word for word in words if word not in stopwords]
    
    return words
```

```python
# applying the function on the title feature:

df['title'] = df['title'].apply(title_token)
```

```python
# creating a function which can concatinate the text(list of words of paras) and title(list of words of title) together:

def create_tags(list_of_wordlist, title):
    
    list_of_wordlist.append(title)
    
    return list_of_wordlist
```

```python
# applying the function on the dataframe using lambda function:

df['tags'] = df.apply(lambda row: create_tags(row['text'], row['title']), axis=1)
```

## Text Analysis

```python
# importing the positive and negative words text files.:

positive = "C:\\Users\\dk957\\Downloads\\positive-words.txt"
negative = "C:\\Users\\dk957\\Downloads\\negative-words.txt"

# Open the file in read mode
with open(positive, 'r') as file:
    positive_words = file.read().split()
    
with open(negative, 'r') as file:
    negative_words = file.read().split()
```

### POSITIVE SCORE

This score is calculated by assigning the value of +1 for each word if found in the Positive Dictionary and then adding up all the values.

```python
# creating a function to calculate positive score:

def positive_points(list_of_listwords):
    
    points = 0
    
    for i in list_of_listwords:
        for j in i:
            if j in positive_words:
                points += 1
            else:
                continue
                
    return points
```

```python
df['POSITIVE SCORE'] = df['tags'].apply(positive_points)
```

### NEGATIVE SCORE

This score is calculated by assigning the value of -1 for each word if found in the Negative Dictionary and then adding up all the values. We multiply the score with -1 so that the score is a positive number.

```python
# creating a function to calculate negative score:

def negative_points(list_of_listwords):
    
    points = 0
    
    for i in list_of_listwords:
        for j in i:
            if j in negative_words:
                points -= 1
            else:
                continue
                
    return points*-1
```

```python
df['NEGATIVE SCORE'] = df['tags'].apply(negative_points)
```

### POLARITY SCORE

Polarity Score = (Positive Score – Negative Score)/ ((Positive Score + Negative Score) + 0.000001)

Range is from 0 to +1

```python
# calculating polarity score using for each row using lambda function:

df['POLARITY SCORE'] = df.apply(lambda row: (row['POSITIVE SCORE']-row['NEGATIVE SCORE'])/((row['POSITIVE SCORE']+row['NEGATIVE SCORE'])+ 0.000001),axis=1)
```

```python
df.head()
```

### SUBJECTIVITY SCORE

Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)

Range is from 0 to +1

```python
# creating a function to calculate subjectivity score:

def subjectivity(tags_of_words,positive_score,negative_score):
    
    total_words = 0
    
    for i in tags_of_words:
        for j in i:
            total_words += 1
    
    score = (positive_score+negative_score)/((total_words)+0.000001)
    return score
```

```python
# calculating subjectivity score by applying the created function using lambda function:

df['SUBJECTIVITY SCORE'] = df.apply(lambda x: subjectivity(x['tags'],x['POSITIVE SCORE'],x['NEGATIVE SCORE']), axis=1)
```

### AVERAGE SENTENCE LENGTH

Average Sentence Length = the number of words / the number of sentences

```python
# creating a function to calculate average sentence length:

def average_sentence_len(tags):
    
    words = 0
    
    for i in tags:
        words += len(i)
             
    return words/len(tags)
```

### PERCENTAGE OF COMPLEX WORDS

Percentage of Complex words = the number of complex words / the number of words

```python
# creating a function to manually calculate the count of syllables in a given word:

def count_syllables(word):
    vowels = "aeiou"
    count = 0
    prev_char_is_vowel = False

    # Function to check if a character is a vowel
    def is_vowel(char):
        return char.lower() in vowels

    for char in word.lower():
        if is_vowel(char):
            if not prev_char_is_vowel:
                count += 1
            prev_char_is_vowel = True
        else:
            prev_char_is_vowel = False

    # Adjust count for words ending with 'e'
    if word.lower().endswith('e') and count > 1 and not word.lower().endswith(('es', 'ed')):
        count -= 1

    return max(1, count)
```

```python
# creating a function to identify the complex words using above function:

def complex_words(text):
    """
    Identify English words in the text that have more than two syllables.
    """
    words = text.split()
    english_words_more_than_two_syllables = [word for word in words if count_syllables(word) > 2]
    return english_words_more_than_two_syllables
```

```python
# creating a function to find complex words percentage:

def complex_words_percentage(tags):
    
    words = 0
    comp_words = 0
    
    for i in tags:
        words += len(i)
        comp_words = comp_words + len(complex_words(" ".join(i)))
             
    return comp_words/words
```

```python
# applying the function to calculate percentage of complex words for each row:

df['PERCENTAGE OF COMPLEX WORDS'] = df['tags'].apply(complex_words_percentage)
```

### AVG NUMBER OF WORDS PER SENTENCE

Average Number of Words Per Sentence = the total number of words / the total number of sentences

```python
# creating a function to calcualate averge number of words per sentence:

def average_words(tags):
    
    words = 0
    sentences= 0
    
    for i in tags:
        words += len(i)
        sentences += 1
        
    return words/sentences
```

### WORD COUNT

We count the total cleaned words present in the text by:
1. removing the stop words (using stopwords class of nltk package).
2. removing any punctuations like ? ! , . from the word before counting

```python
# creating a function to count the total words in each tag:

def total_words(tags):
    
    words = 0
    
    for i in tags:
        words += len(i)
        
    return words
```

### SYLLABLE PER WORD

We count the number of Syllables in each word of the text by counting the vowels present in each word. We also handle some exceptions like words ending with "es","ed" by not counting them as a syllable.

```python
# creating a function to calculate the average syllable count per word in a tag:

def syllables_per_word(tags):
    
    syll = 0
    
    for i in tags:
        for j in i:
            syll += count_syllables(j)
            
    return syll/total_words(tags)
```

```python
# applying the function on the tags feature:

df['SYLLABLE PER WORD'] = df['tags'].apply(syllables_per_word)
```

### PERSONAL PRONOUNS

To calculate Personal Pronouns mentioned in the text, we use regex to find the counts of the words - “I,” “we,” “my,” “ours,” and “us”. Special care is taken so that the country name US is not included in the list.

```python
# creating a function to count the personal pronouns in each tag:

def personal_pronouns(tags):
    
    count = 0
    
    for i in tags:
        for j in i:
            if j in ["i","we","my","ours","us"]:
                count += 1
            else:
                continue
                
    return count
```

```python
# applying the function on the tags feature:

df['PERSONAL PRONOUNS'] = df['tags'].apply(personal_pronouns)
```
